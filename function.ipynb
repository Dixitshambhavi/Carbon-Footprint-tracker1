```python
import pandas as pd

# Load the Excel file (adjust path if needed)
file_path = r"C:\Users\asus\Downloads\carbon_footprint_tracker_multiuser_20users.xlsx"
df = pd.read_excel(file_path)

# Display top rows
df.head()
```

```python
#Total Emissions per User
user_totals = df.groupby("UserID")["CO2e (kg)"].sum().sort_values(ascending=False)
print(user_totals)
```

```python
#Emissions Over Time
date_totals = df.groupby("Date")["CO2e (kg)"].sum()
date_totals.plot(kind="line", title="Total Emissions Over Time")
```

```python
# Category-Wise Emissions

category_totals = df.groupby("Category")["CO2e (kg)"].sum()
category_totals.plot(kind="bar", title="Emissions by Category")
import matplotlib.pyplot as plt
plt.show()
```

```python
# Total CO2 Emissions per User

user_emissions = df.groupby("UserID")["CO2e (kg)"].sum().sort_values(ascending=False)
print(user_emissions.head())

#  Total by Category

category_emissions = df.groupby("Category")["CO2e (kg)"].sum()
print(category_emissions)

#Daily Total Emissions (All Users)
daily = df.groupby("Date")["CO2e (kg)"].sum()

```

```python
#User Comparison (Top Emitters)

top_users = user_emissions.head(10)
top_users.plot(kind="bar", title="Top 10 Emitting Users")
plt.ylabel("CO2e (kg)")
plt.show()
```
